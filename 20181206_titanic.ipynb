### 1. 일변량(Uni-variate) 질적 자료의 분석
* 표
  * 빈도
  * 백분율
* 그래프
  * 막대그래프
  * 원그래프

일변량이란 하나의 열(변수), 하나의 열을 잘 분석하는 것도 상관분석, 회귀분석 등에 비해 중요
* 질적인 자료 > 표와 그래프
* 양적인 자료 > 표와 그래프

등등 데이터의 질/양에 대한 분석 후 분석 방법을 정할 수 있다.

````python
# 빈도(Frequency)
# data["variable"].value_counts()
# data.variable.value_counts()
# 두가지 방법 모두 가능
df_train.sex.value_counts()
````

````python
df_train.pclass.value_counts()
````

````python
df_train.pclass.value_counts(ascending=True) # 빈도 기준 오름차순
````

````python
df_train.pclass.value_counts().sort_index() # 인덱스 순서
````

````python
# 백분율(Percent) : (빈도/합계) * 100
# 백분율의 소수점은 최종적으로 하나만 가짐(default)
# data["variable"].value_counts(normalize = True)*100 # 변수 여러개, for문 사용할 때 좋다.
# data.variable.value_counts(normalize = True)*100 # 변수 하나
df_train.sex.value_counts(normalize=True)*100
````

````python
# 반올림
# round(숫자, ndigits = 반올림할자리수숫자)
round(df_train.sex.value_counts(normalize=True)*100, ndigits = 2)
````

````python
import matplotlib.pyplot as plt
import platform

# 그래프 한글 깨짐 방지
if platform.system() == 'Darwin':
    # 맥일때
    plt.rcParams["font.family"] = 'AppleGothic'
elif platform.system() == 'Windows':
    # 윈도우일때
    plt.rcParams["font.family"] = 'Malgun Gothic'

# 마이너스기호(-) 폰트 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False

# 간혹 그래프가 주피터 노트북 내에서 보여지지 않을 때, 잘 보여지면 주석 풀 필요 없음
#%matplotlib inline
````

````python
# 막대그래프
# plt.bar(lable, frequency or percent)
# plt.show() # 메시지 없이 그림만 보이게
sex_frequency = df_train.sex.value_counts() # 빈도 기준 내림차순으로 결과 저장
plt.bar(sex_frequency.index, sex_frequency)
plt.show()
````

````python
sex_frequency = df_train.sex.value_counts() # 빈도 기준 내림차순으로 결과 저장
plt.bar(sex_frequency.index, sex_frequency)
plt.title("성별빈도")
plt.xlabel("성별")
plt.ylabel("빈도")
plt.show()
````

````python
# 원그래프
sex_percent = round(df_train.sex.value_counts(normalize=True)*100, ndigits = 2)
plt.pie(sex_percent, labels = sex_percent.index)
plt.show()
````
